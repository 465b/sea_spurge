```python
import numpy as np
import os

import matplotlib.pyplot as plt
from oceantracker.read_output.python import load_output_files
```

```python
def load_chunked_data(path_to_dir, key):
    # stats data is seperated into the different cases
    # hence, we concat them back together
    
    # find all sucessfull runs
    dirs = [os.path.join(path_to_dir,d) for d in os.listdir(path_to_dir) if os.path.isdir(os.path.join(path_to_dir,d))]
    case_info_files = []
    for d in dirs:
        case_info = [f for f in os.listdir(d) if f.endswith('_caseInfo.json')]
        case_info_files.append(os.path.join(d, case_info[0]))
    case_info_files = sorted(case_info_files)

    case_info = load_output_files.read_case_info_file(os.path.join(path_to_dir,case_info_files[0]))

    stats = case_info['output_files']['particle_statistics']
    print(stats.keys())


    connectivity_matrix = []
    for case_info_path in case_info_files:
        connectivity_chunk = load_output_files.load_stats_data(case_info_path, key)
        connectivity_matrix.append(connectivity_chunk['connectivity_matrix'])

    connectivity_matrix = np.concatenate(connectivity_matrix, axis=1)

    return connectivity_matrix
```
